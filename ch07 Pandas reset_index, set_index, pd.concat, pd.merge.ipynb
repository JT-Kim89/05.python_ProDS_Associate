# 01 데이터 병합 개요

## 병합의 종류
- 데이터를 이어 붙이는 바인딩(Binding)은 행 또는 열 방향으로 연산 가능
- 특정 변수를 기준으로 두 데이터를 엮는 조인(Join)을 많이 사용
- 조인 연산 중 Inner Join과 Left Join을 많이 사용

## 조인(Join) 상세
- Inner Join: 두 데이터 세트의 기준이 되는 변수인 Key 변수가 공통인 행을 반환
- Left Join: A세트의 모든 행을 유지하고 B세트는 Key 변수 기준으로 A세트와 공통인 행을 반환

# 02 주요 함수 및 메서드 소개
## pandas - reset_index()
- 기존의 인덱스를 초기화하는 메서드
- 데이터프레임의 인덱스를 초기화할 경우 0부터 row개수 만큼의 수사자가 재할당

```python
import pandas as pd
```

```python
bike = pd.read_csv("강의자료/실습파일/bike.csv")
bike.head(2)
```

```python
bike_sub = bike.sample(n = 4, random_state = 123)
bike_sub
```

```python
bike_sub.reset_index() #새로운 인덱스가 생김
```

```python
bike_sub = bike_sub.reset_index(drop = True)
bike_sub
```

## pandas - set_index()
- 특정 변수를 인덱스로 지정할 경우 사용하는 메서드
- 데이터 병합 또는 시계열 분해에서 연산을 위해 활용

```python
bike_sub = bike_sub.set_index("datetime")
bike_sub
```

```python
bike_sub.reset_index()
```

## pandas - concat()
- 데이터를 row 방향 또는 column 방향으로 이어 붙이는 함수
- axis 인자에 1을 입력하는 경우 column 방향으로 병합
- 데이터 병합 시 서로 다른 변수 속성 또는 인덱스에 주의

```python
bike_1 = bike.iloc[:3, :4]
bike_1
```

```python
bike_2 = bike.iloc[5:8, :4]
bike_2
```

```python
pd.concat([bike_1, bike_2])
```

```python
pd.concat([bike_1, bike_2], axis =1)  # index 번호가 달라서 따로 연결됨
```

```python
pd.concat([bike_1, bike_2.reset_index(drop = True)],
          axis =1)
```

```python
pd.concat([bike_1.reset_index(drop = True),
           bike_2.reset_index(drop = True)],
          axis =1)
```

## pandas - merge()
- 데이터프레임을 변수 기준으로 병합할 때 사용하는 함수
- inner join, left join 등 다양한 join 연산이 가능하며 비슷한 함수로는 join이 있음
- left_on과 right_on 인자에 각 데이터의 key 변수를 지정

```python
df_A = pd.read_csv("강의자료/실습파일/join_data_group_members.csv")
df_B = pd.read_csv("강의자료/실습파일/join_data_member_room.csv")
```

```python
df_A
```

```python
df_B
```

```python
pd.merge(left = df_A, right = df_B,
        left_on = "member", right_on = "name",
        how = "inner")
```

```python
pd.merge(left = df_A, right = df_B,
        left_on = "member", right_on = "name",
        how = "outer")
```

## Q2 여름과 겨울의 시간대별 registered 평균을 비교할 때 가장 차이가 많이 나는 시각은?

```python
Q2 = pd.read_csv("강의자료/실습파일/bike.csv")
Q2.head(2)
```

```python
Q2["datetime"] = pd.to_datetime(Q2["datetime"])
Q2.info()
```

```python
Q2["hour"] = Q2["datetime"].dt.hour
Q2.head()
```

```python
Summer = Q2[Q2["season"] == 2]
Winter = Q2[Q2["season"] == 4]
```

```python
print(Summer["season"].unique())
print(Winter["season"].unique())
```

```python
Summer_mean = Summer.groupby("hour")["registered"].mean()
Winter_mean = Winter.groupby("hour")["registered"].mean()
Ans = abs(Summer_mean - Winter_mean)
Ans
```

```python
Ans.idxmax()
```

```python
# concat 활용 정답
Season_mean = pd.concat([Summer_mean, Winter_mean], axis = 1)
# Season_mean = Season_mean.iloc[:, [0, 1, 3]]   # reset_index를 활용한 경우 추출
Season_mean
```

```python
Season_mean.columns = ["Summer", "Winter"]
Season_mean.head(2)
```

```python
Season_mean["diff"] = Season_mean["Summer"] - Season_mean["Winter"]
Season_mean.head(2)
```

```python
Season_mean.iloc[[Season_mean["diff"].idxmax()]]
```

## Q3 비가 온 날에 30도가 넘는 시각의 count의 평균은 얼마인가?
1) humidity의 최대값이 100인 날을 비가 온 날로 간주한다.

```python
Q3 = pd.read_csv("강의자료/실습파일/bike.csv")
Q3.head(2)
```

```python
Q3["datetime"] = pd.to_datetime(Q3["datetime"])
Q3.info()
```

```python
Q3["date"] = Q3["datetime"].dt.date
Q3.head(2)
```

```python
humax = Q3.groupby(["date"])["humidity"].max().reset_index()
humax
```

```python
rain = humax[humax["humidity"] == 100]
rain.head()
```

```python
Q3_rain = pd.merge(left = Q3, right = rain,
                  left_on = "date", right_on = "date",
                  how = "inner")
Q3_rain.head()
```

```python
Q3_rain_T30 = Q3_rain[Q3_rain["temp"] > 30]
Q3_rain_T30.head()
```

```python
Q3_rain_T30["count"].mean()
```
